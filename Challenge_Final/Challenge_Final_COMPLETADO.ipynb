### Remover las filas indicadas (en el Challenge) de los catálogos de tmodulo.

```python
ruta_catalogos_tmodulo = 'conjunto_de_datos_enif_2021_csv/conjunto_de_datos_tmodulo_enif_2021/catalogos'
```

```python
def remover(numero_opcion, lista, ruta_catalogos_tmodulo = ruta_catalogos_tmodulo):
  """
  Dependiendo la opcion, la función recibe una lista de los archivos (cadenas sin extensión csv)
  en las que removerá las filas que lleven por nombre la opción escogida así como la ruta de la carpeta
  de tmodulo en donde se encuentran los catalogos:

  numero_opcion 0 para: Remover Meses
  numero_opcion 1 para: Remover Ingresos
  numero_opcion 2 para: Remover Minutos
  numero_opcion 3 para: Remover Numero de veces
  numero_opcion 4 para: Remover Horas

  y dependiendo cada caso REGRESA un Diccionario de data frames en los que las llaves son cada una
  de las cadenas en las listas y el valor es el df con la fila removida. Para llamar al df se usa
  'cadena mod'
  """
  if numero_opcion in list(range(0,5)): #Se comprueba si la opcion es existente
    dic_elem_rem = {} #Se inicializa el diccionario que contendrá a ls dfs con las filas removidas

    #Ciclo en el que se leé cada csv y remueve las filas, guardando los df's
    for file in lista:
      ruta_archivo = '/' + file + '.' +'csv'
      full_path = ruta_catalogos_tmodulo + ruta_archivo
      #full_path = os.path.join(ruta_catalogos_tmodulo,ruta_archivo)
      df = construye_df(full_path)

      #Para las opciones 0,2,4 se remueven simpre las filas con índice 0
      if numero_opcion in [0,2,4]:
        df.drop(index=0, inplace=True)
      #Para las opciones 1 y 3 se remueven simpre las filas con índice
      else:
        df.drop(index=1, inplace=True)

      df.reset_index(drop = True, inplace = True)
      dic_elem_rem[file] = df

    return dic_elem_rem #Regresa el diccionario con los df's

  else:
    return print("Error, marca una opción correcta")
```

```python
def sustituye_val_num(ruta_catalogos, df_a_cambiar, lista_nombres_cat_cambiados_suma, df_edad_cambiado = edad_tsdem ):
  """Esta función toma CUALQUIERA de las dos rutas hacias los catálogos de tmodulo ó tsdem, uno de estos df,
  una lista con todos los catálogos que previamente se les quitaron filas con la función 'remover' y el primer
  df que cambiamos: edad_tsdem (edad_tmodulo tambien funciona).
  
  Retorna una copia de tmodulo o tsdem con los valores categóricos en vez de numéricos de todos sus 
  catálogos a los que no se les quitaron filas, excepto edad_tsdem. 
  Nota: los valores numéricos de tsdem quedan completamente sustituidos por los categóricos.
  """

  copia = df_a_cambiar.copy()
  ruta_completa = os.path.join(ruta_catalogos,'*.csv')
  #Primero modificar con lo no existentes
  for file in glob(ruta_completa):
    base_name_f = os.path.basename(file)
    label = base_name_f.split(".")[0] #Ayuda a obtener la etiqueta
    if label in lista_nombres_cat_cambiados_suma:
      continue
    df = construye_df(file)
    dic =  dict(df.values)
    dic_cambios = {int(key): value for (key,value) in dic.items()}
    copia[[label.upper()]] = copia[[label.upper()]].replace(dic_cambios)
  dic_edad = dict(df_edad_cambiado.values)
  dic_cambia_edad = {int(key): value for (key,value) in dic_edad.items()}
  copia[["EDAD"]] = df_a_cambiar[["EDAD"]].replace(dic_cambia_edad)

  return copia
```

```python
def completa_sustitución_tmodulo(tmodulo_cambiado, lista_nombres_cat_cambiados,lista_dic_catalogos_cambiados):
  """Esta función termina de sustituir los valores numéricos de tmodulo por sus valores categóricos, complementa
  lo hecho por la función previa "sustituye_val_num".
  Recibe tmodulo_cambiado, lista de los nombres de los catálogos a los que se les quitaron las filas al principio
  y una lista con los diccionarios que nos regresó la función 'remover'.
  
  Retorna una copia de tmodulo_cambiado con todos los valores categóricos
  """
  longitud = len(lista_dic_catalogos_cambiados)
  for i in list(range(longitud)):
    for j in list(range(len(lista_nombres_cat_cambiados[i]))):
      dic_i = lista_dic_catalogos_cambiados[i]
      dic_i_j = dic_i[lista_nombres_cat_cambiados[i][j]]
      #print(lista_nombres_cat_cambiados[i][j])
      dic = dict(dic_i_j.values)
      dic_cambios = {int(key): value for (key,value) in dic.items()}
      tmodulo_cambiado[[lista_nombres_cat_cambiados[i][j].upper()]] = tmodulo_cambiado[[lista_nombres_cat_cambiados[i][j].upper()]].replace(dic_cambios)

  return tmodulo_cambiado
```

### Ahora con los índices podemos generar una función que nos construya un df que contenga un rango salarial 
semanal, quincenal, mensual o anual

```python
lista_tiempos = [tmodulo_cambiado_completo["P3_8B"].unique()[1],tmodulo_cambiado_completo["P3_8B"].unique()[2],tmodulo_cambiado_completo["P3_8B"].unique()[3],tmodulo_cambiado_completo["P3_8B"].unique()[4]]
lista_tiempos
```

```python
lista_indices_ingr = [indices_semana, indices_quincena, indices_mes, indices_ano]
```

```python
def ingresos_por_tiempo(num_tiempo,  df_origin, lista_tiempos = lista_tiempos, lista_indices_ingr = lista_indices_ingr):
  """
  Recibe un número para crear un rango de ingresos por el tiempos deseado:
  1 Crea un rango de ingresos por semana
  2 Crea un rango de ingresos por quincena
  3 Crea un rango de ingresos por mes
  4 Crea un rango de ingresos por año
  
  REGRESA un df con los precios convertidos al periodo de tiempo que se escoja
   """
  lis_num_sem = [2, 4, 50.4]
  lis_num_quin = [2, 2, 24 ]
  lis_num_mes = [4, 2, 12]
  lis_num_ano = [50.4, 24, 12]
  df = df_origin.copy()
  for j in [1,2,3,4]:
    if (num_tiempo == 1) and (j != 4) :
      df[['P3_8B']] = df[['P3_8B']].replace(lista_tiempos,lista_tiempos[0])
      df.loc[lista_indices_ingr[j],['P3_8A']] = df.loc[lista_indices_ingr[j],['P3_8A']].divide(lis_num_sem[j-1])

    if (num_tiempo == 2) and (j != 2):
      df[['P3_8B']] = df[['P3_8B']].replace(lista_tiempos,lista_tiempos[1])
      if j==1:
        df.loc[lista_indices_ingr[j-1],['P3_8A']] = df.loc[lista_indices_ingr[j-1],['P3_8A']].mul(lis_num_quin[j-1])
      else:
        df.loc[lista_indices_ingr[j-1],['P3_8A']] = df.loc[lista_indices_ingr[j-1],['P3_8A']].divide(lis_num_quin[j-2])

    if (num_tiempo == 3) and (j != 3):
      df[['P3_8B']] = df[['P3_8B']].replace(lista_tiempos,lista_tiempos[2])
      if j!=4:
        df.loc[lista_indices_ingr[j-1],['P3_8A']] = df.loc[lista_indices_ingr[j-1],['P3_8A']].mul(lis_num_mes[j-1])
      else:
        df.loc[lista_indices_ingr[j-1],['P3_8A']] = df.loc[lista_indices_ingr[j-1],['P3_8A']].divide(lis_num_mes[j-2])

    if (num_tiempo == 4) and (j != 4):
      df[['P3_8B']] = df[['P3_8B']].replace(lista_tiempos,lista_tiempos[3])
      df.loc[lista_indices_ingr[j-1],['P3_8A']] = df.loc[lista_indices_ingr[j-1],['P3_8A']].mul(lis_num_ano[j-1])

  return df 


```

# Función generadora de gráficas
Observamos que para la mayoría de gráficas restantes es hacer el mismo procedimiento que el que se acaba de hacer.

```python
def genera_grafica(lista_columnas, titulo, x_label, y_label):
    """Esta función toma la lista de las columnas de las que se obtienen la información, el título de la gráfica
    así como las etiquetas en los ejes X e Y.

RETORNA una gráfica con las características antes mencionadas
    """
        
    #Con este ciclo obtenemos las razones (y las contamos) de cada columna
    etiquetas = []
    conteo_etiquetas = []
    j = 0
    m = 0
    for i in lista_columnas:
        """Obtenemos las etiquetas de las columnas así como su frecuencia"""
        etiqueta = tmodulo_cambiado_completo[i][(tmodulo_cambiado_completo[i] != 'No se declaró como opción afirmativa')].dropna().unique()[0]
        etiquetas.append(etiqueta)
        conteo_etiquetas.append(tmodulo_cambiado_completo[i][(tmodulo_cambiado_completo[i] != 'No se declaró como opción afirmativa')].count())
        if etiqueta == 'Otros bancos':
            j = 1
        if etiqueta == 'Otros bancos o instituciones financieras':
            m = 1
    ind_otro = etiquetas.index('Otro')
    Sustitucion = ['Otra institución Bancaria']
    
    """Las condiciones de abajo suman las etiquetas de 'Otro' con 'Otro bancos' ó ''Otros bancos o instituciones financieras'
    (suponemos que eran el mismo dato) dependiendo el caso, se quitan las dos columnas sumadas y se reemplazan
    con una única columna (Sustitucion)"""
    if j == 1:
        ind_otr_bank = etiquetas.index('Otros bancos')
        Suma = [conteo_etiquetas[ind_otro] + conteo_etiquetas[ind_otr_bank]]
        etiquetas.pop(ind_otro)
        etiquetas.pop(ind_otr_bank-1)
        conteo_etiquetas.pop(ind_otro)
        conteo_etiquetas.pop(ind_otr_bank-1)
        etiquetas = etiquetas + Sustitucion
        conteo_etiquetas = conteo_etiquetas + Suma

    if m == 1:
        ind_otr_inst_bank = etiquetas.index('Otros bancos o instituciones financieras')
        Suma = [conteo_etiquetas[ind_otro] + conteo_etiquetas[ind_otr_inst_bank]]
        etiquetas.pop(ind_otro)
        etiquetas.pop(ind_otr_inst_bank - 1)
        conteo_etiquetas.pop(ind_otro)
        conteo_etiquetas.pop(ind_otr_inst_bank - 1)
        etiquetas = etiquetas + Sustitucion
        conteo_etiquetas = conteo_etiquetas + Suma
        
    print(etiquetas)
    print(conteo_etiquetas)
    _, ax_plot = plt.subplots()
    ax_plot.barh(etiquetas, conteo_etiquetas, align='center') 
    ax_plot.set_ylabel(y_label)
    ax_plot.invert_yaxis()  # labels read top-to-bottom
    ax_plot.set_xlabel(x_label)
    ax_plot.set_title(titulo)
    plt.show()
```

# ¿Qué forma de pago utiliza con más frecuencia cuando realiza compras?

Esta información la podemos consultar en las columnas "P7_1_1" (500 ó menos) y "P7_1_2" (500 ó más)

La siguiente función nos ayuda a visaualizar las siguientes dos gráficas

```python
def plotea_forma_pago(columna, observar_sin_etiqueta, ylabel,xlabel,titulo):
    
    """La función toma una columna de forma de pago (P7_1_1 o P7_1_2 o inclusive CUALQUIERA), 
    etiquetas para el eje x, Y , el título y una etiqueta opcional para quitarla de la gráfica, si quiere 
    ver toda la gráfica sin intrrumpciones ponga cualquier cosa.
    Regresa una gráfica con los datos anteriores
    """
    lista_f_pago =[]
    cuenta_f_pago = []
    for i in np.arange(9):
        etiqueta = tmodulo_cambiado_completo[columna].unique()[i] #Etiquetas
        if etiqueta == observar_sin_etiqueta: #podemos observar la gráfica sin alguna etiqueta
            continue
        lista_f_pago.append(etiqueta) #Guardamos las etiquetas
        #Contamos las etiquetas
        cuenta_f_pago.append(tmodulo_cambiado_completo[[columna]][tmodulo_cambiado_completo[columna] == etiqueta].count()[0])
    
    _, ax_plot = plt.subplots()
    ax_plot.barh(lista_f_pago, cuenta_f_pago,  align='center') 
    ax_plot.set_ylabel(ylabel)
    ax_plot.invert_yaxis()  # labels read top-to-bottom
    ax_plot.set_xlabel(xlabel)
    ax_plot.set_title(titulo)
    plt.show()
```
