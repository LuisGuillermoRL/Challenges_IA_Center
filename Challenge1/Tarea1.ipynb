###Creation of the DataFrame

```python
""" 
Each element of the string is put into each corresponding list
"""
year = []
month = []
day = []
decimal_date = []
est_sunspot_numb = []
est_stand_dev = []
num_stations_calcul = []

#loops through each row of the df placing each string in its corresponding list
for i in range(0,75086):
    year.append(int(datos.iloc[i,0].split()[0]))
    month.append(int(datos.iloc[i,0].split()[1]))
    day.append(int(datos.iloc[i,0].split()[2]))
    decimal_date.append(float(datos.iloc[i,0].split()[3]))
    est_sunspot_numb.append(int(datos.iloc[i,0].split()[4]))
    est_stand_dev.append(float(datos.iloc[i,0].split()[5]))
    num_stations_calcul.append(int(datos.iloc[i,0].split()[6]))
```

We create two separate DataFrame's, one with the data corresponding to the date (df_fecha_mod) and the other with the remaining data (datos_frame).

```python
#list(zip()) create the columns corresponding to the DataFrame datos_frame
datos_frame = pd.DataFrame(list(zip(decimal_date,est_sunspot_numb,est_stand_dev,num_stations_calcul)), columns = [' Decimal date',"Estimated Sunspot Number","Estimated Standard Deviation","Number of Stations calculated"])
```

function zip() https://micro.recursospython.com/recursos/la-funcion-zip.html

```python
datos_frame
```

```python
datos_frame.info()
```

```python
#list(zip()) create the columns corresponding to the DataFrame df_fecha
df_fecha = pd.DataFrame(list(zip(year, month,day)), columns = ['year','month','day'])
```

We use the method to_datetime from pandas https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html

```python
#Method to_datetime
df_fecha_mod = pd.DataFrame({"year_month_day":pd.to_datetime(df_fecha)})
```

```python
df_fecha_mod
```

```python
df_fecha_mod.info()
```

###We obtain the complete and clean DataFrame by concatenating the two DataFrame's obtained previously

```python
data_frame = pd.concat([df_fecha_mod,datos_frame], axis=1) #Method concat (horizontal concatenation)
data_frame
    
```

###We save the DataFrame in a csv file

```python
data_frame.to_csv("sunspots.csv", index = False)
```
