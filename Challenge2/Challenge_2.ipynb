###First Part

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data from Puebla
url1 = 'https://www.inegi.org.mx/app/api/denue/v1/consulta/BuscarEntidad/todos/21/1/500/b3c999c7-9584-7861-b677-4131ed9d2bfb'
```

```python
#Import data
df_json = pd.read_json(url1)
```

```python
df_json #DataFrame
```

```python
df_json.columns #DataFrame's Columns
```

```python
df_json[["Ubicacion"]] #Column of interest
```

```python
#separate the elements of the column by commas
ubicaciones = df_json["Ubicacion"].str.split(",", expand = True) 
ubicaciones
```

```python
ubicaciones.columns = ['Localidad', 'Municipio', 'Estado'] #rename the columns
```

```python
df_json_ubicaciones = pd.concat([ubicaciones, df_json], axis = 1) #concatenate the two df
df_json_ubicaciones
```

```python
df_json_ubicaciones.columns #DataFrame's columns
```

```python
df_json_ubicaciones.drop("Ubicacion", axis = 1, inplace = True) #Drop the column of interest: 'Ubicacion'
```

```python
df_json_ubicaciones.columns #DataFrame's columns
```

###Gráfica

```python
#we sort in descending order and obtain the first 10 elements of the "Clase_actividad" column
df_plot = df_json_ubicaciones["Clase_actividad"].value_counts(sort = True).head(10)
df_plot #Panda Series
```

```python
df_plot.index
```

Gráfica

```python
f_plot, ax_plot = plt.subplots()

Clases = df_plot.index
y_pos_plot = np.arange(10)
Numero_de_clases = df_plot

ax_plot.barh(y_pos_plot, Numero_de_clases, align='center') # ([x1,y1],[x2,y2],...,[xn,yn])
ax_plot.set_yticks(y_pos_plot, labels = Clases)
ax_plot.invert_yaxis()  # labels read top-to-bottom
ax_plot.set_xlabel('Número de establecimientos por clase')
ax_plot.set_title('Cantidades por cada clase de actividad Económica en Puebla')

plt.show()
```
